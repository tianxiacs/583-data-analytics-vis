```python
import pandas as pd
import altair as alt
```

```python
# 
# This section is to prepare 2 dataframes for Sketch #1
# Ping Tian if this breaks
# 
dataRQ1 = pd.read_excel('Data_Simple.xlsx', usecols="A, B, E:J");
# Add a column of index for Sketch#1
dataRQ1['index'] = dataRQ1.reset_index().index
dataRQ1 = dataRQ1.rename(columns={'Type (1-5)': 'Type'})
dataRQ1a = dataRQ1.copy()
dataRQ1b = dataRQ1.copy()
# dataRQ1a["Type"] = pd.to_numeric(dataRQ1a["Type"])

# Make two dataframes for scatter plots of feelings before and after to combine into a compound graph later
dataRQ1a['Type'] = dataRQ1a['Type'].apply(lambda x: x-0.075)
dataRQ1b['Type'] = dataRQ1b['Type'].apply(lambda x: x+0.075)
# Append $ to the price after converting float to string
dataRQ1a['Price'] = dataRQ1a['Price'].astype(str)
dataRQ1a['Price'] = dataRQ1a['Price'].apply(lambda x: "$ "+x)
dataRQ1b['Price'] = dataRQ1b['Price'].astype(str)
dataRQ1b['Price'] = dataRQ1b['Price'].apply(lambda x: "$ "+x)
# Add a column with all the detailed health type in text instead of 1-5
dataRQ1a['Healthy Type'] = dataRQ1['Type']
dataRQ1b['Healthy Type'] = dataRQ1['Type']
dataRQ1a['Healthy Type'] = dataRQ1a['Healthy Type'].apply(lambda x: "1 - Healthy" if x == 1 else ("2 - Mostly Healthy" if x == 2 else ("3 - Middle" if x == 3 else ("4 - Mostly Unhealthy" if x == 4 else "5 - Unhealthy"))))
dataRQ1b['Healthy Type'] = dataRQ1b['Healthy Type'].apply(lambda x: "1 - Healthy" if x == 1 else ("2 - Mostly Healthy" if x == 2 else ("3 - Middle" if x == 3 else ("4 - Mostly Unhealthy" if x == 4 else "5 - Unhealthy"))))

```

# Eating Foods on a Scale of 1-5

While exploring ideas for our topic, we collectively agreed that food plays an important role in our lives. It provides the nutrients we need to grow and survive, and in this day and age, it is also a source of pleasure and comfort. Here we asked the question, "What impact does our feelings have on the type of food we eat and the feelings we have after"? Through this research question, we looked to answer what sort of relationship exists between our emotions and the level of food we decide to consume. As simple as it may seem, we all participate in eating multiple times a day. By recording our self-evaluated emotions before and after eating, as well as the type of food we ate, we were able to graph this data into the graph you see below. The type of food is defined on a scale between 1-5 where 1 represents healthy food, and 5 is unhealthy. This is further defined in our previous phases where we listed some of our most commonly eaten foods on this scale to refer back to while recording data. This data is valuable to then reflect on the habits we may have formed in relation to our emotions at the time. 

```python
# 
# This section is to make Sketch #1
# Ping Tian if this breaks
# 

elements1 = ['#b1c000', '#048026', '#c5013a', '#f55600', '#6aa705', '#530659', '#3c003d', '#269242', '#e7d400', '#509600', '#007bd8']
elements2 = ['#f87200', '#e8d300', '#fce8ad', '#f1e543', '#007bd8', '#efdf0b']

elements = elements1 + elements2

dropdown_box = alt.binding_select(options=[None,'Brian Du','Mackenzie Tam','Cassidy Creasy','Tian Xia'])
selection = alt.selection_single(fields=['Person'], bind=dropdown_box, name='Filter')

# Make a compound graph consisting of two scatter plots; one shows the feeling before, and the other shows the feeling after.
before = alt.Chart(dataRQ1a).mark_square(size=500, stroke='black', strokeWidth=0.5).encode(x=alt.X('Type', title='Type: Healthy - 1, Mostly healthy - 2, Middle - 3, Mostly unhealthy - 4, Unhealthy - 5', axis=alt.Axis(tickCount=4), scale=alt.Scale(domain=[0.5, 5.5])), y=alt.Y('index', axis=None), tooltip=['Name of Food', 'Datetime', 'Person', 'Source of food', 'Healthy Type', 'Price', 'Feelings Before', 'Feelings After'], color=alt.condition(selection, 'Feelings Before', alt.value('lightgray'))).add_selection(selection)
after = alt.Chart(dataRQ1b).mark_square(size=500, stroke='black', strokeWidth=0.5).encode(x=alt.X('Type', axis=alt.Axis(tickCount=4, gridColor="#6E85B2", gridWidth=2), scale=alt.Scale(domain=[0.5, 5.5])), y=alt.Y('index', axis=None), tooltip=['Name of Food', 'Datetime', 'Person', 'Source of food', 'Healthy Type', 'Price', 'Feelings Before', 'Feelings After'], color=alt.condition(selection, 'Feelings After', alt.value('lightgray')))

sketch1 = alt.layer(before, after).properties(width=750, height=500, title=['The Type Of Food We Eat And The Influence It Has On Our Emotions']).interactive().configure_title(font='Arial', color='#000000', fontSize=22).configure_view(strokeWidth=4, fill='#FBF3E4', stroke='#C1AC95').configure_legend(labelFont='Arial', titleFontSize=11, labelFontSize=16).configure_axis(labelFontSize=15, titleFontSize=15).configure_range(category= elements)

sketch1
```

This visual is a great tool in drilling down to identify individual habits as well as a broad overview of what the four of us tend to eat relative to our emotions. # Routine Eating

Throughout a busy day of work many people look forward to breakfast, lunch, and dinner. Breakfast signals the start of the day; lunch is that nice break right in the middle and dinner is when you can wind down and relax. Many have set times they eat these meals throughout the day, but sometimes unforeseen circumstances make it difficult to do so. Especially as university students our sleep, work and class schedules are very chaotic. Sometimes we find ourselves working till 2 am and then eating dinner and heading to bed. Some people struggle to find time to eat proper meals at proper times throughout the day. With mental health becoming a popular topic, we wanted to see how mealtimes affect mental health. So, we decided to research how our feelings (before and after) differ from our feelings eating at irregular times. When we have to eat at erratic times throughout our weeks how does that affect our mental health for the better or worse?

My group consisting of Brian, Tian, Cassidy, and I collected data for a week to answer this data. We collected when we ate, what we ate, how healthy the food we ate was, the price of the food, and our feelings before and after eating. For our feelings we used the Robert Plutchik Wheel of Emotions. For health of food, we came up with a scale before hand (1 being the healthiest, and 5 being unhealthy). After the week, we then went over as a group each item eaten and its health rating to ensure there was no bias with the recording. With this data we have created this visualization to answer the question: How do feelings from eating at regular mealtimes differ from the feelings for eating at irregular times?

```python
# 
# This section is to prepare 2 dataframes for Sketch #10
# Ping Tian if this breaks
# 
dataRQ3 = pd.read_excel('Data_Simple.xlsx', usecols="A:J");
dataRQ3 = dataRQ3.rename(columns={'Type (1-5)': 'Type'})
dataRQ3a = dataRQ3.copy()
dataRQ3b = dataRQ3.copy()
# Format Date to only day
dataRQ3a['Date'] = dataRQ3a['Date'].dt.day
dataRQ3b['Date'] = dataRQ3b['Date'].dt.day
# Format Time to float in 24-hour format 
dataRQ3a['Time'] = dataRQ3a['Time'].apply(lambda x: (x.hour + (x.minute/60)))
dataRQ3b['Time'] = dataRQ3b['Time'].apply(lambda x: (x.hour + (x.minute/60)))
# Make two dataframes for scatter plots of feelings before and after to combine into a compound graph later
dataRQ3a['Date'] = dataRQ3a['Date'].apply(lambda x: x-0.07)
dataRQ3b['Date'] = dataRQ3b['Date'].apply(lambda x: x+0.07)
# Append $ to the price after converting float to string
dataRQ3a['Price'] = dataRQ3a['Price'].astype(str)
dataRQ3a['Price'] = dataRQ3a['Price'].apply(lambda x: "$ "+x)
dataRQ3b['Price'] = dataRQ3b['Price'].astype(str)
dataRQ3b['Price'] = dataRQ3b['Price'].apply(lambda x: "$ "+x)
# Add a column with all the detailed health type in text instead of 1-5
dataRQ3a['Healthy Type'] = dataRQ3['Type']
dataRQ3b['Healthy Type'] = dataRQ3['Type']
dataRQ3a['Healthy Type'] = dataRQ3a['Healthy Type'].apply(lambda x: "1 - Healthy" if x == 1 else ("2 - Mostly Healthy" if x == 2 else ("3 - Middle" if x == 3 else ("4 - Mostly Unhealthy" if x == 4 else "5 - Unhealthy"))))
dataRQ3b['Healthy Type'] = dataRQ3b['Healthy Type'].apply(lambda x: "1 - Healthy" if x == 1 else ("2 - Mostly Healthy" if x == 2 else ("3 - Middle" if x == 3 else ("4 - Mostly Unhealthy" if x == 4 else "5 - Unhealthy"))))
```

```python
# 
# This section is to make Sketch #10
# Ping Tian if this breaks
# 

elements1 = ['#b1c000', '#048026', '#c5013a', '#f55600', '#6aa705', '#530659', '#3c003d', '#269242', '#e7d400', '#509600', '#007bd8']
elements2 = ['#f87200', '#e8d300', '#fce8ad', '#f1e543', '#007bd8', '#efdf0b']

elements = elements1 + elements2

dropdown_box2 = alt.binding_select(options=[None,'1 - Healthy','2 - Mostly Healthy','3 - Middle','4 - Mostly Unhealthy', '5 - Unhealthy'])
selection2 = alt.selection_single(fields=['Healthy Type'], bind=dropdown_box2, name='Filter')

# Make a compound graph consisting of two scatter plots; one shows the feeling before, and the other shows the feeling after.
before10 = alt.Chart(dataRQ3a).mark_circle(size=550, stroke='black', strokeWidth=0.5).encode(x=alt.X('Date', title='October', axis=alt.Axis(tickCount=7, gridColor="#A09e9a", gridWidth=2), scale=alt.Scale(domain=[12, 21])), y=alt.Y('Time', title='Time (24-Hour)', axis=alt.Axis(gridColor="#A09e9a", gridWidth=2)), tooltip=['Name of Food', 'Datetime', 'Person', 'Source of food', 'Healthy Type', 'Price', 'Feelings Before', 'Feelings After'], color=alt.condition(selection, 'Feelings Before', alt.value('lightgray'))).add_selection(selection2)
after10 = alt.Chart(dataRQ3b).mark_circle(size=550, stroke='black', strokeWidth=0.5).encode(x=alt.X('Date', axis=alt.Axis(tickCount=7, gridColor="#C1AC95", gridWidth=2), scale=alt.Scale(domain=[12, 21])), y=alt.Y('Time', title='Time in 24-Hour Format', axis=alt.Axis(gridColor="#C1AC95", gridWidth=2)), tooltip=['Name of Food', 'Datetime', 'Person', 'Source of food', 'Healthy Type', 'Price', 'Feelings Before', 'Feelings After'], color=alt.condition(selection, 'Feelings After', alt.value('lightgray')))

sketch10 = alt.layer(before10, after10).properties(width=750, height=500, title=['The Affects Of Eating Food At Different Times Of Day']).interactive().configure_title(font='Arial', color='#000000', fontSize=35).configure_view(strokeWidth=4, fill='#F9F9F9', stroke='#B5CDA3').configure_legend(labelFont='Arial', titleFontSize=10, labelFontSize=16).configure_axis(labelFontSize=15, titleFontSize=15).configure_range(category= elements)

sketch10
```

Here we have time on the y-axis in 24-hour format, and the days we recorded on the x-axis. Each circle represents a feeling. The circle on the left of a line represents the feelings before eating and the circle on the right of the line represents the feeling after eating. Color coded on the side is our legend showing what color each emotion is represented as. We selected the same colors as the ones used in Robert Plutchiks Wheel of Emotions to maintain consistency. Using the hover function on the circles, you can see all of the information: when we ate, what we ate, and our feelings before and after eating. Zooming out the circles begin to overlap so using the zoom function you can see more specifically what times we ate and our feelings before and after. With our filter we can see the health of the food we ate (1 being healthy, and 5 being unhealthy).
